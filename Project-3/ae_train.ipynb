```python
import os
import pickle
from zipfile import ZipFile
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf 
from autoencoder import Autoencoder
import librosa.display
```

```python
with ZipFile('/content/train_test_data.zip', 'r') as z:
    z.extractall()
```

```python
x_train = np.load('/content/train_test_data/x_train.npy')
y_train = np.load('/content/train_test_data/y_train.npy')
x_test = np.load('/content/train_test_data/x_test.npy')
y_test = np.load('/content/train_test_data/y_test.npy')
```

```python
def create_ae(num_layers, latent_dim):
    if num_layers == 2:

        ae = Autoencoder(
            input_shape=(256, 64, 1),
            conv_filters=(100, 50),
            conv_kernels=(3, 3),
            conv_strides=(2, 2),
            latent_space_dim=latent_dim
        )   
    elif num_layers == 3:
        ae = Autoencoder(
            input_shape=(256, 64, 1),
            conv_filters=(150, 100, 50),
            conv_kernels=(3, 3, 3),
            conv_strides=(2, 2, 2),
            latent_space_dim=latent_dim
        )   

    else:
        raise ValueError('only 2 or 3 layers allowed')

    return ae
```

##RANDOM SEARCH IN THE HYPERPARAMETERS

```python
lr, b, eps = 0.0005, 32, 50
save_dir = '/content/drive/MyDrive/models/denoising_autoencoder/'

depth = [2, 2, 3, 3]
dim = [32, 64, 32, 64]
```

```python
i = 0

ae = create_ae()
model_dir = os.path.join(save_dir, f'{depth[i]}_{dim[i]}')
ae.compile(learning_rate=lr)
ae.denoise_train(x_train, y_train, batch_size=b, num_epochs=eps)
ae.save(model_dir)
```
